```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from datetime import datetime
```

```python
df = pd.read_csv('dataset-16d-14d.csv')
df.head()
```

```python
unix_list = []
for index in range(len(df)):
    date = str(df['time'][index]).split('.')[0].rstrip('Z')
    date_format = datetime.strptime(date, "%Y-%m-%dT%H:%M:%S")
    unix_time = datetime.timestamp(date_format)
    unix_list.append(unix_time)

df['unix'] = pd.Series(unix_list)

for index in range(len(df)):
    time = str(df.loc[index, 'time']).split('.')[0]
    if 'Z' not in time:
        time += 'Z'
    df.loc[index, 'time'] = time

df['time'] = pd.to_datetime(df['time'])
```

```python
scaler = MinMaxScaler()

numerical_columns = []
for col in df.columns:
    if df[col].dtype == 'float64' and col != 'unix':
        numerical_columns.append(col)

scaled_df = df
scaled_df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
scaled_df.head()
```

```python
plt.figure(figsize=(60, 15))
for col in scaled_df.columns[:5]:
    if scaled_df[col].dtype == 'float64':
        if col != 'unix':
            sns.lineplot(data=scaled_df, x='time', y=col, legend='brief', linewidth=1.5)

plt.show()


plt.figure(figsize=(60, 15))
for col in scaled_df.columns[5:]:
    if scaled_df[col].dtype == 'float64':
        if col != 'unix':
            sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=1.5)

plt.show()
```

```python
plt.figure(figsize=(50, 10))
for col in scaled_df.columns[:]:
    if scaled_df[col].dtype == 'float64':
        if col != 'unix':
            if col == 'temperature':
                sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=0.5, color='red')
            elif col == 'axialAxisRmsVibration':
                sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=0.5, color='blue')
            elif col == 'radialAxisKurtosis':
                sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=5.5, color='orange')
            elif col == 'radialAxisPeakAcceleration':
                sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=1.5, color='green')
            elif col == 'radialAxisRmsAcceleration':
                sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=0.5, color='yellow')
            elif col == 'radialAxisRmsVibration':
                sns.lineplot(data=scaled_df, x='time', y=col, legend='full', linewidth=0.5, color='purple')
plt.show()
```

```python
mask = np.zeros_like(df[numerical_columns].corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True

plt.figure()
sns.heatmap(df[numerical_columns].corr(), mask = mask, annot = True, annot_kws = {"size":10})
sns.set_style("white")
plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)
plt.show()
```
